```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# choix du p et q et d=0 car pas de différenciation.
p_range = range(0, 3)  #par exemple on cherche p de 0 à 2
q_range = range(0, 3)

best_aic = float("inf")
best_params = None

for p in p_range:
    for q in q_range:
        try:
            modele_arima = sm.tsa.ARIMA(serie_temporelle, order=(p, 0, q))  # d=0
            resultats = modele_arima.fit()
            aic = resultats.aic
            if aic < best_aic:
                best_aic = aic
                best_params = (p, 0, q)  # d=0
        except:
            continue

print(f"Meilleurs paramètres (p, d, q) : {best_params}")
print(f"Meilleur AIC : {best_aic}")
```
